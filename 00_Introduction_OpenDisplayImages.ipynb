```python
from matplotlib import pyplot as plt
import cv2

print( cv2.__version__)
```

## Loading images

You can use `cv2.imread()` to load an image file.

```python
image = cv2.imread("images/rgb.jpg")

width, height, channels = image.shape

print('%d by %d pixels and %d channels' % (height, width, channels))
```

## Displaying images

You can use `matplotlib.pyplot` (referenced here as `plt`) to display the image. Note that cv2 orders the color channels blue, green then read and matplotlib uses red, green then blue. To address this you can use `cv2.cvtColor()` to change the order before display.

You can use `axis()` and `title()` methods of pyplot to adjust the appearance.

```python
imageRGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display original image using matplotlib
plt.imshow(image)
plt.axis('off')
plt.title('Incorrect (BGR image)')
plt.show()

# Display converted image using matplotlib
plt.imshow(imageRGB)
plt.axis('off')
plt.title('Correct (RGB image)')
plt.show()
```

## Loading images from URL

The openCV `cv2.imread()` method is able to read directly from a URL, but scikit-image (skimage) libraries `io.imread()` method is. This method load an image in RGB channel order so no conversion is needed.

```python
from skimage import io

url = 'https://github.com/Algorithmic-Lens/Learning-image-manipulation-in-Python/raw/master/images/rgb.jpg'
remoteImage = io.imread(url)

# Display image using matplotlib
plt.imshow(remoteImage)
plt.axis('off')
plt.title('Correct (RGB image)')
plt.show()
```
